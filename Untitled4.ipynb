```python
data["High"].plot(figsize=[15,8])
```

```python
# lets take the data for certain limit ie from 2022 to 2024 and for price also we will be taking from 0 to 410 

data["High"].plot(xlim=['2022-01-01','2024-03-14'],ylim=[0,410],c="red",figsize=[15,8])
```

```python
data=data.loc["2022-01-01":"2024-03-14"]  # taking the data from 2022 till now
```

```python
stock_open=data.loc["2022-01-01":"2024-03-14"]["Open"] 
# seeing the stock open data from 2022 to till now
stock_open
```

```python
index=data.index
index
```

```python
# plotting the stock_open data
fig,ax=plt.subplots()
ax.plot(index,stock_open)
plt.tight_layout()
fig.autofmt_xdate() #prevents the overlapping
```

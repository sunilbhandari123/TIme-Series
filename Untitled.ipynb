lets no analysis the data by visiulizing we have two variable ie carbon monoxide concentration and relative-humidity

```python
for var in ['CO_sensor','RH']:
    data[var].plot(figsize=(20,4))
    plt.title(var)
    plt.ylabel(var)
    plt.show()

```

```python
# here we can see the long tail in downward direction ie the values are negative but the values of the 
#carbon monoxide and the relative humidity cannot be negative so it is outiler
```

```python
# removing negative values of co and rh 
data=data.loc[(data["CO_sensor"]>=0) & (data["RH"]>=0)]
#just neglection the negative values
```

```python
for var in ['CO_sensor','RH']:
    data[var].plot(figsize=(20,4))
    plt.title(var)
    plt.ylabel(var)
    plt.show()
```

In this data there might be the missing value while taking the time stamp and we mightnot see by our eye
also so we will be marking it

```python
# let see for 1 hour time stamp
data_=data.asfreq('1H')

#lets apply fordward fill method.
data_imputed=data_.fillna(method='ffill')

#lets plot imputed data
for var in ['CO_sensor','RH']:
    ax=data_imputed[var].plot(linestyle="-", marker=".",figsize=(20,4))
    
    #lets plot the imputed value on the top in red
    
    data_imputed[data_.isnull()][var].plot(ax=ax,legend=None,color='red',marker='.')
    
    plt.title(var)
    plt.show()
```

Now lets see Seasonality ie there is high concentration of CO and RH when there is lots of traffic in peak hours

```python
#lets see in hourly basis extracting the hours form the time stamp

time_=data.index.time
time_=pd.Series(time_,index=data.index)
time_.head()
```

```python
#ploting the avarage concentration per hour

data.groupby(time_)[["CO_sensor"]].mean().plot(figsize=(8,4))
plt.title("Concentration Seasonality per hour")
plt.show()
```

as we can see the concentration is lowest at morning/ night and highest at the peak hours that means there
is seasonality

```python
data.groupby(time_)[['RH']].mean().plot(figsize=(8,4))
plt.title('Concentration Seasonality of RH per hours')
plt.show()
```

As we can see the relative humidity is highest at morning and night and decreases at day time.

Feature Engineering

```python
data.head()
```

```python
 data.sort_index(inplace=True)
```

```python
#lets extract the more feature from the following data

data["Day"]=data.index.day
data["Month"]=data.index.month
data["Week"]=data.index.isocalendar().week
data["Day_of_week"]=data.index.day_of_week
data["Hour"]=data.index.hour
data['is_Weekend']=np.where(data["Day_of_week"]>4,1,0)
```

```python
data.head()
```

finding the lag feature let do it for 1 hours and 24 hours based on that we will be predicting the
concentration of next hours.

```python
variables=['CO_sensor','RH']
#shift the data fordward 1 hour
tempo=data[variables].shift(freq='1H')
#new variable names
tempo.columns=[v +'lag_1' for v in variables]
#merge
data=data.merge(tempo,left_index=True,right_index=True,how='left')
```

```python
data.head()
```

```python
data[['CO_sensor','CO_sensorlag_1']].head()
```

we can see there are missing values lets check for it

```python
data.isnull().sum()
```

now lets see the lags for 24 hours

```python
tempo=data[variables].shift(freq="24H")
tempo.columns=(v +"lag24" for v in variables)
data=data.merge(tempo,left_index=True,right_index=True,how="left")
```

```python


data[["CO_sensor", "CO_sensorlag24"]].head(25)
```

 Windows feature..
we will take the mean of the last 3 values

```python
tempo=(
    data[variables].rolling(window="3H").mean()# avarage of last 3 hours
    .shift(freq="1H")#move the avarage 1 hour fordward
)
tempo.columns=[v+'windows' for v in variables]
```

```python
tempo.head()
```

```python
#merging with orginal variables
data.merge(tempo,left_index=True,right_index=True,how="left")
```

Perodic Features

```python
cyclical = CyclicalFeatures(
    variables=["Month", "Hour"],  # The features we want to transform.
    drop_original=False,  # Whether to drop the original features.
)

data = cyclical.fit_transform(data)
```
